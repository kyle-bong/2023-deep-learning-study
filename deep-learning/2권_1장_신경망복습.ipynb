##1.1.4 벡터의 내적과 행렬의 곱

벡터의 내적은 두 벡터에서 대응하는 원소들의 곱을 모두 더한 것

★벡터의 내적은 직관적으로는 '두 벡터가 얼마나 같은 방향을 향하고 있는가를 나타냅니다. 벡터의 길이가 1인 경우로 한정하면, 완전히 같은 방향을 향하는 두 벡터의 내적은 1이 됩니다. 반대로 반대 방향을 향하는 두 벡터의 내적은 -1입니다.

구체적으로 말하면, 두 벡터 A = [A1, A2, ..., An]와 B = [B1, B2, ..., Bn]의 내적은 A1B1 + A2B2 + ... + An*Bn으로 계산됩니다. 만약 두 벡터가 정규화된 경우 (각 벡터의 길이가 1인 경우), 그들의 내적은 두 벡터 사이의 각도 코사인에 해당합니다.

image.png

이 그림에서, 빨간색과 파란색 벡터는 각각 벡터 A와 B를 나타냅니다. 이 두 벡터의 내적은 빨간색 벡터의 길이와 파란색 벡터의 길이를 곱한 값에 빨간색 벡터와 파란색 벡터 사이의 각도(θ)의 코사인 값을 곱한 값입니다.

만약 두 벡터가 완전히 같은 방향을 가리키면 (즉, 각도가 0도라면), 코사인 값은 1이 되므로 내적은 두 벡터의 길이 곱과 같아집니다. 반면, 두 벡터가 완전히 반대 방향을 가리키면 (즉, 각도가 180도라면), 코사인 값은 -1이 되어 내적은 두 벡터의 길이 곱의 반대 값이 됩니다. 두 벡터가 서로 수직일 경우 (즉, 각도가 90도라면), 코사인 값은 0이 되므로 내적은 0이 됩니다.

이처럼 내적은 두 벡터의 방향성과 길이에 대한 정보를 포함하며, 이는 두 벡터의 유사도(similarity)를 측정하는 데 매우 유용한 도구입니다. 특히, 자연어 처리에서는 단어 간의 유사도를 측정하기 위해 벡터의 내적을 사용하는 경우가 많습니다.

입력층(Input Layer):

출력층(Output Layer):

활성화 함수(Activation Function):

 활성화 함수는 뉴런의 출력을 결정하는 함수입니다. 예를 들어, 시그모이드, ReLU, tanh 등이 있습니다.

## 1.2.2 계층으로 클래스화 및 순전파 구현

입력 X가 Affine 계층, Sigmoid 계층, Affine 계층을 거쳐 결과 S를 출력하는 순전파 구현.
image.png

```python
class Sigmoid:

    def __init__(self):
        self.params = []

    def forward(self, x):
        return 1/(1+np.exp(-x))

class Affine:

    def __init__(self, W, b):
        self.params = [W, b]

    def forward(self, x):
        W, b = self.params
        out = np.matmul(x, W) + b
        return out
```

```python
class TwoLayerNet:
    def __init__(self, input_size, hidden_size, output_size):
        I, H, O = input_size, hidden_size, output_size

        # 가중치와 편향 초기화
        W1 = np.random.randn(I, H)
        b1 = np.random.randn(H)
        W2 = np.random.randn(H, O)
        b2 = np.random.randn(O)

        #계층 생성
        self.layers = [
            Affine(W1, b1),
            Sigmoid(),
            Affine(W2, b2)
        ]

        # 모든 가중치를 리스트에 모은다.
        self.params = []
        for layer in self.layers:
            self.params += layer.params

    def predict(self, x):
        for layer in self.layers:
            x = layer.forward(x)
        return x
```

```python
x = np.random.randn(10, 3)
model = TwoLayerNet(3, 4, 3)
s = model.predict(x)

s
```

## 1.3.1 손실 함수
신경망 학습에 학습이 얼마나 잘 되고 있는지를 알기 위한 '척도'가 필요합니다.
일반적으로 신경망의 성능을 나타내는 척도로 "손실"을 사용함.
손실은 신경망의 예측 결과와 정답 데이터를 비교하여 예측이 얼마나 나쁜가를 산출하는 값을 사용함.

다중 클래스 분류 신경망에서는 손실 함수로 흔히 "교차 엔트로피 오차"를 이용함.

앞 서 구성한 신경망에서 손실 함수를 구현하기 위해, Softmax 계층과 Cross Entropy Error 계층을 새로 추가 합니다. (Softmax 계층은 소프트맥스 함수를, Cross Entropy Error 계층은 교차 엔트로피 오차를 구하는 계층 입니다.

image.png

x = 입력 데이터

t = 정답 레이블

L = 손실

```python
class Softmax:
    def __init__(self):
        self.params = []

    def forward(self, x):
        if x.ndim == 2:
            x = x - x.max(axis=1, keepdims=True)
            x = np.exp(x)
            x /= x.sum(axis=1, keepdims=True)
        elif x.ndim == 1:
            x = x - np.max(x)
            x = np.exp(x) / np.sum(np.exp(x))

        return x

class CrossEntropyError:
    def __init__(self):
        self.params = []

    def forward(self, y, t):
        if y.ndim == 1:
            t = t.reshape(1, t.size)
            y = y.reshape(1, y.size)

        if t.size == y.size:
            t = t.argmax(axis=1)

        batch_size = y.shape[0]
        return -np.sum(np.log(y[np.arange(batch_size), t] + 1e-7)) / batch_size
```

```python
class TwoLayerNet:
    def __init__(self, input_size, hidden_size, output_size):
        I, H, O = input_size, hidden_size, output_size

        # 가중치와 편향 초기화
        W1 = np.random.randn(I, H)
        b1 = np.random.randn(H)
        W2 = np.random.randn(H, O)
        b2 = np.random.randn(O)

        # 계층 생성
        self.layers = [
            Affine(W1, b1),
            Sigmoid(),
            Affine(W2, b2),
            Softmax()
        ]

        self.loss_layer = CrossEntropyError()

        # 모든 가중치와 기울기를 리스트에 모은다.
        self.params = []
        for layer in self.layers:
            self.params += layer.params

    def predict(self, x):
        for layer in self.layers:
            x = layer.forward(x)
        return x

    def loss(self, x, t):
        y = self.predict(x)
        return self.loss_layer.forward(y, t)
```

## 1.3.3 연쇄 법칙 (Chain Rule)

연쇄 법칙은 합성 함수의 도함수를 구하는 방법에 대한 규칙입니다. 합성 함수는 여러 함수의 결합으로 이루어진 함수를 말하며, 간단히 말해서 하나의 함수의 출력이 다른 함수의 입력으로 사용되는 경우, 이들 함수의 결합을 합성 함수라고 합니다.

예를 들어, 두 함수 f(x)와 g(x)가 있고, h(x) = f(g(x))와 같이 합성된 함수 h(x)가 있다고 가정해봅시다. 이때, h'(x)는 다음과 같이 계산될 수 있습니다.

h'(x) = f'(g(x)) * g'(x)

이것이 바로 연쇄 법칙의 핵심입니다. 여기서 f'(g(x))는 f의 g(x)에서의 미분을 나타내고, g'(x)는 g의 x에서의 미분을 나타냅니다.

신경망의 역전파(backpropagation)에서는 이 연쇄 법칙이 핵심적으로 사용됩니다. 각 노드를 통과하며 기울기(미분값)을 계산하고, 이를 통해 오차를 입력층까지 전달하여 각 가중치를 업데이트하는 방식에서 연쇄 법칙이 필요합니다.

## 덧셈노드, 곱셈 노드, 분기 노드, Reapeat 노드, Sum 노드, Matmul 노드

1. 덧셈 노드 (Addition Node): 두 개의 입력 값을 더하여 출력 값으로 반환합니다. 이 노드에서의 미분은 1입니다.

2. 곱셈 노드 (Multiplication Node): 두 개의 입력 값을 곱하여 출력 값으로 반환합니다. 곱셈 노드에서의 미분은 상대적인 입력 값에 의존합니다. 즉, x에 대한 미분은 y, y에 대한 미분은 x입니다.

3. 분기 노드 (Branch Node): 하나의 입력 값을 두 개 이상의 출력 값으로 복사합니다. 이 노드에서의 미분은 입력 값에 무관하게 출력 값의 합으로 계산됩니다.

4. Repeat 노드: 하나의 입력 값을 복사하여 여러개의 출력 값으로 반환합니다. 역전파시에는 모든 기울기의 합을 구합니다.

5. Sum 노드: 여러 입력 값을 모두 더하여 하나의 출력 값으로 반환합니다. 역전파시에는 입력의 수에 상관 없이 상류에서 전해진 기울기를 그대로 하류로 전달합니다.

6. Matmul 노드 (Matrix Multiplication Node): 두 개의 입력 행렬을 곱하여 출력 행렬을 반환합니다. 이 노드에서의 미분은 행렬 곱셈의 규칙을 따르며, 상류의 기울기와 자신의 입력을 이용해 기울기를 계산합니다.

## 1.3.5 기울기 도출과 역전파 구현

1. Sigmoid 계층 구현

image.png

```python
class Sigmoid:
    def __init__(self):
        self.params, self.grads = [], []
        self.out = None

    def forward(self, x):
        out = 1 / (1 + np.exp(-x))
        self.out = out
        return out

    def backward(self, dout):
        dx = dout * (1.0 - self.out) * self.out
        return dx
```

2. Affine 계층 구현

image.png

```python
class Affine:
    def __init__(self, W, b):
        self.params = [W, b]
        self.grads = [np.zeros_like(W), np.zeros_like(b)]
        self.x = None

    def forward(self, x):
        W, b = self.params
        out = np.matmul(x, W) + b
        self.x = x
        return out

    def backward(self, dout):
        W, b = self.params
        dx = np.matmul(dout, W.T)
        dW = np.matmul(self.x.T, dout)
        db = np.sum(dout, axis=0)

        self.grads[0][...] = dW
        self.grads[1][...] = db
        return dx
```

```python
def load_data(seed=np.random.randint(1000, 10000)):
    np.random.seed(seed)
    N = 100  # 클래스당 샘플 수
    DIM = 2  # 데이터 요소 수
    CLS_NUM = 3  # 클래스 수

    x = np.zeros((N*CLS_NUM, DIM))
    t = np.zeros((N*CLS_NUM, CLS_NUM), dtype=np.int)

    for j in range(CLS_NUM):
        for i in range(N): #N*j, N*(j+1)):
            rate = i / N
            radius = 1.0*rate
            theta = j*4.0 + 4.0*rate + np.random.randn()*0.2

            ix = N*j + i
            x[ix] = np.array([radius*np.sin(theta), radius*np.cos(theta)]).flatten()
            t[ix, j] = 1

    return x, t
```

```python
x, t = load_data()

# 데이터셋을 그래프로 시각화
N = 100
CLS_NUM = 3
markers = ['o', 'x', '^']
for i in range(CLS_NUM):
    plt.scatter(x[i*N:(i+1)*N, 0], x[i*N:(i+1)*N, 1], s=40, marker=markers[i])
plt.show()
```
